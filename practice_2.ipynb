```python
def dna(data):
    count_a = 0
    count_t = 0
    count_g = 0
    count_c = 0
    for item in data:
        if item == "a":
            count_a += 1
        elif item == "t":
            count_t += 1
        elif item == "g":
            count_g += 1
        elif item == "c":
            count_c += 1
    print(f"""
    a는 {count_a}개
    t는 {count_t}개
    g는 {count_g}개
    c는 {count_c}개""")

data = input("염기서열을 입력하시오: ")
dna(data)
```

```python
print("Player who got Head part wins")
datas = ["Head", "Tail"]
import random
data_1 = random.choice(datas)
data_2 = random.choice(datas)

def coin(data_1, data_2):
    if data_1 == "Head" and data_2 == "Head":
        return "draw"
    elif data_1 == "Tail" and data_2 == "Tail":
        return "draw"
    elif data_1 == "Head" and data_2 == "Tail":
        return "P1 wins"
    elif data_1 == "Tail" and data_2 == "Head":
        return "P2 wins"

x = input("If you want to start a game, enter 'yes': ")
if x == 'yes':
    n = input("P1 enter 'go' to toss a coin: ")
    if n == 'go':
        print("P1: ", data_1)
    n = input("P2 enter 'go' to toss a coin: ")
    if n == 'go':
        print("P2: ", data_2)
    print(coin(data_1, data_2))
```
